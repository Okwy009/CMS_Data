## RN (Registered Nurse) Hours

```python
# Increase the figure size for better visibility
plt.figure(figsize=(16, 8))

# Plot distribution for RN hours with more bins
sns.histplot(df_cleaned['Hrs_RN'], kde=True, bins=50, color='blue')

# Adjust labels and title for better readability
plt.title('Distribution of RN Hours', fontsize=20)
plt.xlabel('RN Hours', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

# Show the plot
plt.show()
```

## CNA (Certified Nursing Assistant) Hours

```python
# Increase the figure size for better visibility
plt.figure(figsize=(16, 8))

# Plot distribution for CNA hours with more bins
sns.histplot(df_cleaned['Hrs_CNA'], kde=True, bins=50, color='green')

# Adjust labels and title for better readability
plt.title('Distribution of CNA Hours', fontsize=20)
plt.xlabel('CNA Hours', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

# Show the plot
plt.show()
```

## Exploring the Relationships

- Finding correlation between MDScensus (resident census) and staffing hours
- Determines if more residents lead to more staffing hours.

```python
# Check the correlation between census and staffing hours
df_cleaned[['MDScensus', 'Hrs_RN', 'Hrs_CNA']].corr()
```

This table represents the correlation matrix for the variables MDScensus (resident census from MDS), Hrs_RN (total hours for registered nurses), and Hrs_CNA (total hours for certified nursing assistants). The values in the matrix indicate how these variables are correlated with each other:

1. MDScensus and Hrs_CNA (0.8949): There is a strong positive correlation between the resident census and the hours worked by CNAs. This suggests that as the number of residents in a nursing home increases, the CNA hours also increase. It indicates a direct relationship between resident care needs and staffing levels.

2. MDScensus and Hrs_RN (0.5725): There is a moderate positive correlation between the resident census and the hours worked by registered nurses. This means that nursing homes with more residents tend to have higher RN hours, but the correlation is not as strong as it is for CNAs.

3. Hrs_RN and Hrs_CNA (0.6221): There is a moderate positive correlation between RN and CNA hours. This indicates that facilities with higher RN staffing hours also tend to have more CNA staffing hours, reflecting a coordinated staffing strategy.

## Insights

- The strong correlation between MDScensus and Hrs_CNA suggests that CNA staffing is closely tied to the number of residents, possibly indicating that CNAs are the primary staff involved in direct resident care.
- The moderate correlation between MDScensus and Hrs_RN suggests that RN staffing might be influenced by other factors in addition to the resident census, such as administrative tasks or facility regulations.
- Understanding these correlations is essential for Clipboard Health, as it can help tailor staffing solutions based on nursing home size and resident needs.
